2 Divide the data with the following portions:
   - Training data 70 %
   - Test data 30 %

```python
import pandas as pd 
from sklearn import metrics, tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Load the dataset and save it to the dataframe
data = pd.read_csv('data_files/voice.csv')

# Save features and labels in variables
x = data[["meanfun", "IQR"]]
y = data["label"]

# split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

# Display the splitted data to verify
print("Training data shape:", x_train.shape, y_train.shape)
print("Test data shape:", x_test.shape, y_test.shape)
```

3 Define classifier without the max_depth parameter. Use get_depth() method for the defined classifier. What value for tree depth was chosen automatically?

```python
# Define the classifier for the decision tree
clf = DecisionTreeClassifier( random_state=42)
clf.fit(x_train, y_train)
```

```python
# Get the depth of the decision tree
tree_depth = clf.get_depth()
tree_depth
```

4 Visualise the tree structure and create the confusion matrix where you can compare true and predicted labels.

```python
# save the feature name and label variables 
feature_names = x.columns
labels = y.unique()

labels
```

```python
# Visualize the tree structure
plt.figure(figsize = (16, 10), facecolor = 'lightgray')

tree.plot_tree(clf, filled=True, feature_names = feature_names, class_names = labels, fontsize = 0)
plt.show()
```

```python
# Use predict method for our test and save it to the variables
test_pred_decision_tree = clf.predict(x_test)

# Confusion matrix definition
confusion_matrix = metrics.confusion_matrix(y_test, test_pred_decision_tree)

# Input values for the confusion matrix
ax = plt.axes()
sns.set(font_scale = 1.3)
plt.figure(figsize = (10,7))
sns.heatmap(confusion_matrix, annot=True, fmt = "g", ax = ax, cmap = "magma")

# Add descriptive title for the figure and names for axles
ax.set_title("Confusion Matrix - Decision Tree")
ax.set_xlabel("Predicted label", fontsize = 15)
ax.set_xticklabels(labels)
ax.set_ylabel("True Label", fontsize = 15)
ax.set_yticklabels(list(labels), rotation = 0)
plt.show()
```

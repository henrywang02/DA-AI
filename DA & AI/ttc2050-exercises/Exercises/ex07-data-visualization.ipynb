1 Read the earthquakes.csv file and save its content into a variable. Then use bar chart to visualize magnitude values being 8.2 or over so that date values will be presented in x-axle and magnitude in y-axle.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the earthquake.csv file and save its content unto a variable
data = pd.read_csv('data_files/earthquakes.csv')


# Filter earthquakes with magnitude 8.2 or over
filtered_data = data[data['Magnitude'] >= 8.2]

# Plotting the bar chart
filtered_data.plot(kind="bar", x="Date", y="Magnitude", figsize=(12, 6), color="green")
plt.title('Earthquakes based on Magnitude of 8.2 or over')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

2 Read the gamedata.json file to a new variable. Visualize the data with pie chart so that share count will be counted for each Steam rating (in other words how many percentage out of all games have received which rating, for example Very Positive rating).

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# First we need to load the data

with open('data_files/gamedata.json', 'r') as file:
    game_data = json.load(file)

# data checked
print(game_data)
    
steam_rating=[entry['steamRatingText'] for entry in game_data]
steam_count = {rating: steam_rating.count(rating) for rating in set(steam_rating)}

# Ploting with Json

labels = steam_count.keys()
size = steam_count.values()

plt.figure(figsize=(8,8))
# plt.legend(game_data['steamRatingText'])
plt.pie(size, autopct='%1.1f%%', colors=['violet','lightcoral', 'lightskyblue','red', 'gray', 'green'])
# Add legend
plt.legend(labels, title='Steam Ratings', loc='upper right')
plt.title('Distribution of Steam Ratings')
plt.show()

```

3 Use the same game_data.json as above and create a scatter plot where you compare the relationships for metacriticScore and salePrice.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the gamedata.json file to a new variable
data = pd.read_json('data_files/gamedata.json')

# Plotting the Scatter Chart
data.plot(kind='scatter', x='metacriticScore', y='salePrice', figsize=(12,6), sizes=(20, 40), alpha=0.5, color='red')
plt.title('The Relationships of metacriticScore and salePrice')
plt.grid(True)
```

4 Use earthquake data you read earlier (the unfiltered original data) and create a scatter plot for magnitude values using Longitude and Latitude values for x- and y-axles correspondingly. In addition, use colormap for magnitude values.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the earthquake data
data = pd.read_csv('data_files/earthquakes.csv')

# Plotting the data as a requirement
data.plot(kind="scatter", x="Longitude", y="Latitude", figsize=(12, 6), c="Magnitude", cmap='turbo', sizes=(20, 200))
plt.title('Earthquake Data (Unfiltered)')
```
